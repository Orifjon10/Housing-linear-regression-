```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
import joblib
```

```python
x = df.drop('median_house_value',axis=1)
```

```python
y = df['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
sc = StandardScaler()
oe = OneHotEncoder()
```

```python
numerical_f = x_train.select_dtypes(include=['float']).columns
categorical_f = x_train.select_dtypes(include=['object']).columns
```

```python
protcessor = ColumnTransformer(
    transformers = [
        ('num', sc, numerical_f),
        ('cat',oe, categorical_f)
    ]
)
```

```python
model = make_pipeline(protcessor,LinearRegression())
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, predictions)
```

```python
print('MSE: ',mse)
```

```python
import numpy as np 
print('MSE: ', np.sqrt(mse))
```
